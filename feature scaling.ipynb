```python
from sklearn.cluster import KMeans
```

```python
wcs=[]
for i in range(1,11):
  km=KMeans(n_clusters=i)
  km.fit_predict(df)
  wcs.append(km.inertia_)
```

```python
wcs
```

```python
plt.plot(range(1,11),wcs)
```

```python
x= df.iloc[:,:].values
km=KMeans(n_clusters=4)
y_means=km.fit_predict(x)
```
